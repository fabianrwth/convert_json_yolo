```python
import numpy as np
import os
import json
import cv2

font = cv2.FONT_HERSHEY_SIMPLEX
font_scale = 1

counter = 0

#Load the json file with the annotation metadata
datapath = './data'
with open(os.path.join(datapath, 'mvtec_screws.json')) as f:
  data = json.load(f)

images = data['images']
annotations = data['annotations']

# Loop through all files in the directory and remove them
for filename in os.listdir('./data/labels'):
    file_path = os.path.join('./data/labels', filename)
    if os.path.isfile(file_path):
        os.remove(file_path)


# Mapping annotations to related image
image_mapping = dict()
for img in images:
    image_mapping[img['id']] = []
    
for idx, annotation in enumerate(annotations):
    img_id = annotation['image_id']
    image_mapping[img_id].append(idx)


# Loop through all images
for image_id in image_mapping.keys():

    print(f"IMAGE ID: {image_id}")
    indices = image_mapping[image_id]


    img_path = os.path.join(datapath,'images', images[image_id-1]['file_name'])

    img = cv2.imread(img_path)
    
 
    # Loop through all annotations of a specific image (image_id)
    for annotation in [annotations[i] for i in indices]:
        img_id = annotation['image_id']
        category = annotation['category_id']
        bbox = annotation['bbox']
        width = images[img_id-1]['width']
        height = images[img_id-1]['height']


        x_center = bbox[1]
        y_center = bbox[0]
        bbox_w = bbox[2]
        bbox_h = bbox[3]
        angle = (- bbox[4] + np.pi) * 180 / np.pi
        
        # calculate the four corners of the rotated box
        box = cv2.boxPoints(((x_center, y_center), (bbox_w, bbox_h), angle))
        box = np.int0(box)

        xy_min = (min(box[:,0]), min(box[:,1]))
        xy_max = (max(box[:,0]), max(box[:,1]))

        # Calculate the width and height of the rectangle
        bbox_width_new = abs(xy_min[0] - xy_max[0])
        bbox_height_new = abs(xy_min[1] - xy_max[1])

        # draw the rotated box on the image
        img = cv2.drawContours(img, [box], 0, (0, 0, 255), 4)
        
        # add the category label next to the box
        # label = "{}".format(category)
        # (label_width, label_height), _ = cv2.getTextSize(label, font, font_scale, 1)
        # label_offset_x = int(x_center - label_width/2)
        # label_offset_y = int(y_center + bbox_h/2 + label_height)
        # cv2.putText(img, label, (label_offset_x, label_offset_y), font, font_scale, (20, 20, 20), thickness=4, lineType=cv2.LINE_AA)

        # cv2.rectangle(img, xy_min, xy_max, (0, 255, 0), 2)

        # YOLO format
        x_center_yolo = x_center / width
        y_center_yolo = y_center / height
        w_yolo = bbox_width_new / width
        h_yolo = bbox_height_new / height
        yolo_list = [category, x_center_yolo, y_center_yolo, w_yolo, h_yolo]
        yolo_list[1:] = [round(x, 4) for x in yolo_list[1:]]
        yolo_string = " ".join(str(x) for x in yolo_list)

        with open(os.path.join(datapath, 'labels', images[image_id-1]['file_name'].replace('.png', '.txt')), 'a') as f:
            f.write(f"{yolo_string}\n")

        #print(yolo_string)
        counter += 1

print(counter)


# cv2.imshow('img', cv2.resize(img, None, fx=0.5, fy=0.5))
# cv2.waitKey(0)
# cv2.destroyAllWindows()
# cv2.imwrite('output.jpg', cv2.resize(img, None, fx=0.5, fy=0.5))
```
